# Introdução

## Variáveis

Regras para criar nomes de variáveis
*   Nomes de variáveis deve iniciar com letra ou `_`
*   Não pode ter espaço

### Exemplo
Aqui está um exemplo simples de como variáveis podem ser usadas em Python:

```python
# Variáveis de diferentes tipos
idade = 25
nome = "Carlos"
altura = 1.75
estudante = True

# Exibindo valores e tipos
print(f"Nome: {nome}, Idade: {idade}, Altura: {altura}, Estudante: {estudante}")
print(type(idade))  # <class 'int'>
print(type(nome))  # <class 'str'>
print(type(altura))  # <class 'float'>
print(type(estudante))  # <class 'bool'>
```

### Estilo (nomenclatura) - criação variáveis (convenção)

No Python você pode escolher a nomenclatura que quiser para as variáveis funções e classes, mas existem algumas convenções que organizam melhor esses nomes.

* **Camel case** - é pouco usado em Python
    Ex: minhaVariavelVariasOutrasPalavras

* **Snake case** - é mais comum em Python
    Ex: minha_variavel_varias_outras_palavras

* **Uppercase** - é mais usado para constantes
    Ex: FATURAMENTO = 1000

* **Capital camel case** - parecido com Camel case, mas começa com maiúsculo. É muito utilizado no Python para escrever Classes
    Ex: MinhaVariavelVariasOutras
    
### Variáveis Globais e Locais
* Globais: Variáveis definidas fora de funções ou blocos de código. Podem ser acessadas em qualquer parte do código.  

* Locais: Variáveis definidas dentro de funções. Só são acessíveis dentro do escopo da função.

## Constante

Em Python, você pode definir constantes usando variáveis, mas seguindo uma convenção de nomenclatura específica, já que **a linguagem não possui uma forma nativa de declarar constantes imutáveis**. A convenção é utilizar letras maiúsculas para nomear variáveis que devem ser tratadas como constantes.

#### Como Definir uma Constante
Por convenção, você define uma constante escrevendo o nome da variável em letras maiúsculas. 

Exemplos:

```python
PI = 3.14159
RADIUS = 5

# Cálculo da área de um círculo
area = PI * RADIUS**2
print("Área do círculo:", area)
```

#### Importando Constantes de Outros Módulos

Se você deseja usar constantes em diferentes arquivos, pode defini-las em um arquivo separado e importá-las. 

Exemplo:

```python
# arquivo constants.py
PI = 3.14159
GRAVITY = 9.81

# arquivo main.py
# from constants import PI, GRAVITY

print("Valor de PI:", PI)
print("Gravidade:", GRAVITY)
```

#### Considerações
Embora as constantes sejam tratadas como tal por convenção, nada impede que elas sejam alteradas em Python. A linguagem não possui mecanismos para garantir a imutabilidade de uma variável como as constantes em outras linguagens de programação. Portanto, é importante seguir as convenções para indicar aos desenvolvedores que uma determinada variável não deve ser modificada.
