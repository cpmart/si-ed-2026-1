## Tipos de Dados em Python

* Importância dos tipos de dados
* Breve visão geral dos principais tipos de dados


### Inteiros (int)

* Definição: Números inteiros sem casas decimais
* Exemplos: `a = 10`, `b = -5`, `c = 0`
* Limitações: Limitado pela memória disponível
* Problemas: Pode afetar desempenho em operações complexas

```python
inteiro = 10
print('inteiro é uma variável do tipo: ',type(inteiro))

# saída:
```

## Ponto Flutuante (float)

* Definição: Números reais com casas decimais
* Exemplos: x = 3.14, y = -0.001
* Limitações: Precisão limitada, problemas de arredondamento
* Problemas: Comparações imprecisas
* Tamanho em Bytes: 24 bytes*

```python
# Fórmula de juros compostos: A = P * (1 + r/n)^(nt)
principal = 1000.0  # Capital inicial
rate = 0.05         # Taxa de juros anual
time = 5            # Tempo em anos

amount = principal * (1 + rate)**time

print("Valor após 5 anos:", amount)
print("A variável amount é do tipo: ", type(amount))

# saída:
```

## Booleano (bool)

* Definição: Valores de verdade, True ou False
* Exemplos: is_active = True, has_permission = False
* Limitações: Apenas dois valores possíveis
* Problemas: Sem problemas significativos
* Tamanho em Bytes: 28 bytes

```python
# Verificando se um usuário tem permissão para acessar uma funcionalidade
is_admin = True
has_access = False

if is_admin or has_access:
    print("Acesso permitido")
else:
    print("Acesso negado")

# saída:
```

## String (str)

* Definição: Sequências de caracteres (texto)
* Exemplos: name = "Alice", greeting = 'Hello, World!'
* Limitações: Limitado pela memória disponível
* Problemas: Manipulação de strings longas pode ser ineficiente
* Tamanho em Bytes: 49 bytes + 1 byte por caractere

```python
# Gerando uma mensagem de boas-vindas personalizada
user_name = "Alice"
welcome_message = f"Bem-vinda, {user_name}!"

print(welcome_message)

# saída:
```

## Listas (list)

* Definição: Coleções ordenadas de elementos
* Exemplos: numbers = [1, 2, 3, 4, 5], mixed = [1, "apple", 3.14, True]
* Limitações: Limitado pela memória disponível
* Problemas: Inserção e remoção podem ser lentas
* Tamanho em Bytes: 64 bytes + 8 bytes por elemento

```python
# Armazenando notas de alunos em uma lista
grades = [85, 92, 78, 90, 88]

average_grade = sum(grades) / len(grades)

print("Média das notas:", average_grade)
```

## Tuplas (tuple)

* Definição: Semelhante a listas, mas imutáveis
* Exemplos: coordinates = (10, 20), colors = ('red', 'green', 'blue')
* Limitações: Não podem ser modificadas após criação
* Problemas: Imutabilidade pode ser uma limitação
* Tamanho em Bytes: 48 bytes + 8 bytes por elemento

```python
# Armazenando coordenadas de localização em uma tupla
location = (40.7128, -74.0060)  # Latitude e Longitude de Nova York
print("Coordenadas de Nova York:", location)
```

## Dicionários (dict)

* Definição: Coleções de pares chave-valor
* Exemplos: person = {"name": "John", "age": 30}, config = {"host": "localhost", "port": 8080}
* Limitações: Chave deve ser imutável
* Problemas: Grandes dicionários podem usar muita memória
* Tamanho em Bytes: 240 bytes para 8 pares chave-valor

```python
# Armazenando informações de um cliente em um dicionário
customer = {
    "name": "Douglas da Silva",
    "age": 30,
    "email": "douglas@fatec.edu",
    "loyalty_points": 120
}

print("Nome do cliente:", customer["name"])
print("Pontos de fidelidade:", customer["loyalty_points"])
```

## Conjuntos (set)

* Definição: Coleções não ordenadas de elementos únicos
* Exemplos: fruits = {"apple", "banana", "cherry"}
* Limitações: Elementos devem ser imutáveis
* Problemas: Ordem dos elementos não garantida
* Tamanho em Bytes: 224 bytes para 3 elementos

```python
# Removendo e-mails duplicados de uma lista usando um conjunto
email_list = ["a@example.com", "b@example.com", "a@example.com", "c@example.com"]
unique_emails = set(email_list)

print("Lista de e-mails única:", unique_emails)
```

```python
A = {1,2,2,3,4,5}
print(A)
B = {3,4,5,6}
print (B - A)  # diferença
print (B | A)  # união
print (B & A)  # interseção
print (2 in A) # pertinência
```
