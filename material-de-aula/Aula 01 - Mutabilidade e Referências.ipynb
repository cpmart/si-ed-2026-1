## 1. Variáveis NÃO armazenam valores

Em Python:

Variável = referência para um objeto


````python
a = 10
print(f"id(10) = {id(10)}")
b = a
print(f"id(a) = {id(a)}")
print(f"id(b) = {id(b)}")
b += 1
print(f"id(b) = {id(b)}")
````

Inteiros são imutáveis → `b` passou a apontar para outro objeto.

````python
x = [1,2,3]
y = x
y.append(4)

print(x)  
````

Listas são mutáveis → o objeto foi alterado.

## 2. Mutável vs Imutável

Imutáveis:
- int
- float
- str
- tuple

Mutáveis:
- list
- dict
- set


## 3. Igualdade vs Identidade

```python
a = [1,2]
b = [1,2]

print(a == b)  # True
print(a is b)  # False
```

==  → mesmo conteúdo  
is  → mesmo objeto
