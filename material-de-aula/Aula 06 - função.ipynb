# Funções

## Exemplos mais simples

### Função para calcular a soma de dois números:

````python
def soma(a, b):
    return a + b

resultado = soma(10, 5)
print(resultado)  # Saída: 15
````

### Função para verificar se um número é par:

````python
def eh_par(numero):
    if numero % 2 == 0:
        return True
    return False

print(eh_par(8))  
print(eh_par(7))  

# Saída:
````

### Função para calcular o fatorial de um número:

````python
def fatorial(n):
    fat = 1
    for i in range(1,n+1):
        fat *= i
    return fat
print(fatorial(5))  # Saída: 120
````

### Função para encontrar o maior número em uma lista:

````python
def maior_numero(lista):
    return max(lista)

numeros = [3, 10, 7, 22, 1]
print(maior_numero(numeros))  
````

### Função que converte uma string para maiúsculas:

````python
def para_maiusculas(texto):
    return texto.upper()

print(para_maiusculas("python é incrível"))  
# Saída:
````

### Função que retorna uma saudação com um nome pré-definido:

````python
def saudacao(nome="Visitante"):
    return f"Olá, {nome}!"
print(saudacao())      # Saída: Olá, Visitante!
print(saudacao("Carlos")) # Saída: Olá, Carlos!
````

### Função para calcular o valor de um produto com desconto (o desconto tem valor padrão de 10%):

````python
def preco_com_desconto(preco, desconto=0.10):
    return preco * (1 - desconto)

print(preco_com_desconto(100))       
print(preco_com_desconto(100, 0.20)) 

# Saída:
````

### Função que retorna uma potência com valor padrão para o expoente:

````python
def potencia(base, expoente=2):
    return base ** expoente

print(potencia(3))          
print(potencia(2, 3))       

# Saída:
````

### Função que concatena duas strings, com a segunda string pré-definida como vazia:

````python
def concatenar(str1, str2=""):
    return str1 + str2

print(concatenar("Olá"))        
print(concatenar("Olá", " Mundo!"))

# Saída:
````

### Função para calcular a área de um retângulo, onde um lado tem valor padrão:

````python
def area_retangulo(lado1, lado2=10):
    return lado1 * lado2

print(area_retangulo(5))           
print(area_retangulo(5, 4))        

# Saída:
````

## Exemplos de parâmetros nomeados

### Função com parâmetros nomeados

Neste e xemplo, estamos nomeando os parâmetros largura e altura ao chamar a função, o que torna a chamada mais explícita.

````python
def calcular_area_retangulo(largura, altura):
    return largura * altura

# Chamada usando parâmetros nomeados
resultado = calcular_area_retangulo(altura=10,largura=5)
print(f"Área: {resultado}")  # Saída: Área: 50

# Saída:
````

Aqui, saudacao tem um valor padrão, e podemos passar ou não um valor para ele ao chamar a função.

## Exemplos usando *args:
*args permite passar um número variável de argumentos para uma função. Ele os armazena como uma tupla.

### Função que soma uma quantidade variável de números:

````python
def somar(*args):
    return sum(args)

print(somar(1, 2, 3))  
print(somar(4, 5, 6, 7))  

# Saída: 
````

### Função que imprime todos os argumentos passados:

````python
def imprimir_args(*args):
    for arg in args:
        print(arg)

imprimir_args("Python", 42, [1, 2, 3], True)

# Saída:
````

## Exemplos usando **kwargs:
**kwargs permite passar um número variável de argumentos nomeados (como pares chave-valor) para uma função. Ele os armazena como um dicionário.

### Função que imprime as informações de um usuário:

````python
def informacoes_usuario(**kwargs):
    for chave, valor in kwargs.items():
        print(f"{chave}: {valor}")

informacoes_usuario(nome="João", idade=30, cidade="São Paulo")

# Saída:
````

### Função que constrói uma frase a partir de argumentos nomeados:

````python
def construir_frase(**kwargs):
    return f"O {kwargs.get('animal', 'animal')} {kwargs.get('nome', 'sem nome')} é muito {kwargs.get('adjetivo', 'chato')}."

print(construir_frase(animal="gato", nome="Miau", adjetivo="tosco"))

# Saída: 
````

## Combinando *args e **kwargs:

### Função que recebe tanto argumentos posicionais quanto nomeados:

````python
def mostrar_dados(*args, **kwargs):
    print("Args:", args)
    print("Kwargs:", kwargs)

mostrar_dados(1, 2, 3, nome="Alice", idade=25)

# Saída:
````

## Funções Recursivas
Funções recursivas são aquelas que chamam a si mesmas para resolver um problema, geralmente com uma condição de parada para evitar loops infinitos.

### Cálculo do fatorial

Uma função recursiva para calcular o fatorial de um número.

Neste e xemplo, fatorial(5) retorna 5 * fatorial(4), e assim por diante, até que fatorial(1) retorne 1, encerrando a recursão.

````python
def fatorial(n):
    if n == 0 or n == 1:  # Condição de parada
        return 1
    else:
        return n * fatorial(n - 1)

print(fatorial(5))  # Saída: 120
````

### Sequência de Fibonacci
Função recursiva para calcular o n-ésimo número da sequência de Fibonacci, onde:

Fibonacci(0) = 0

Fibonacci(1) = 1

Fibonacci(n) = Fibonacci(n-1) + Fibonacci(n-2) para n > 1

A função chama a si mesma duas vezes até atingir os casos-base (quando n é 0 ou 1).

````python
def fibonacci(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonacci(n - 1) + fibonacci(n - 2)

print(fibonacci(6))  

# Saída:
````

### Soma dos elementos de uma lista

Função recursiva para somar os elementos de uma lista.

Aqui, a função recursiva remove o primeiro elemento da lista e soma com a soma dos elementos restantes.

````python
def somar_lista(lista):
    if len(lista) == 0:  # Condição de parada
        return 0
    else:
        return lista[0] + somar_lista(lista[1:])  # Soma o primeiro elemento e chama a função para o restante

numeros = [1, 2, 3, 4, 5]
print(somar_lista(numeros))  

# Saída:
````

### Potência de um número

Função recursiva para calcular a potência de um número aˆb.

A função recursiva multiplica a base por ela mesma e chama a função novamente com o expoente diminuído até que o expoente seja 0.

````python
def potencia(base, expoente):
    if expoente == 0:  # Condição de parada
        return 1
    else:
        return base * potencia(base, expoente - 1)

print(potencia(2, 3))  

# Saída: 
````

### Funções lambda 

Em Python, uma **função lambda** é uma maneira rápida e simples de criar pequenas funções anônimas (ou seja, funções sem nome) que podem ter qualquer número de argumentos, mas apenas uma expressão. A expressão é avaliada e retornada quando a função lambda é chamada.

A sintaxe básica de uma função lambda é:

```python
lambda argumentos: expressão
```

#### Exemplos de Uso

**Função Lambda Simples**:

````python
dobro = lambda x: x * 2
print(dobro(5))  
   
# Saída:
````

Aqui, `lambda x: x * 2` cria uma função anônima que multiplica o argumento `x` por 2. A função é atribuída à variável `dobro`.

**Lambda com Múltiplos Argumentos**:

````python
soma = lambda x, y: x + y
print(soma(3, 4))  

# Saída:
````

   Neste exemplo, a função lambda recebe dois argumentos (`x` e `y`) e retorna a soma deles.

**Uso de Lambda em Funções como `map` e `filter`**:

````python
lista = [1, 2, 3, 4, 5]
dobrados = list(map(lambda x: x * 2, lista))
print(dobrados)  

# Saída:
````

   Aqui, a função `map` aplica a função lambda a cada elemento da lista, retornando uma nova lista com os valores dobrados.

#### Quando Usar Lambdas

- **Funções Simples e Temporárias**: Lambdas são ideais para funções pequenas e simples que são usadas temporariamente, especialmente em operações como mapeamento, filtragem ou classificação de dados.
- **Código Conciso**: Se você precisar de uma função curta que será usada apenas uma vez ou poucas vezes, lambdas podem tornar o código mais conciso.

#### Limitações

- **Legibilidade**: O uso excessivo de lambdas pode tornar o código menos legível, especialmente se a lógica da função for complexa.
- **Sem Nome**: Como são funções anônimas, lambdas não têm nome, o que pode dificultar a depuração.

Funções lambda são uma ferramenta poderosa e flexível em Python, especialmente quando usadas em conjunto com outras funções de ordem superior como `map`, `filter` e `sorted`.

### Exemplos de uso de funções lambda combinadas com `print` em Python:

#### **Calcular o Quadrado de um Número**
Usa lambda para calcular e imprimir o quadrado de um número.
