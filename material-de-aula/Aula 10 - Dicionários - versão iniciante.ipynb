**O que é um dicionário?**

Um dicionário guarda informações no formato:

**chave : valor**

Exemplos:

- `"nome": "Ana"`
- `"idade": 19`
- `"cidade": "Jales"`

Pense assim:

- a **chave** é o nome da informação
- o **valor** é o conteúdo daquela informação


## Exemplo visual

```python
aluno = {
    "nome": "Ana",
    "idade": 19,
    "curso": "Engenharia Civil"
}
```

Nesse exemplo:

- `"nome"` é uma chave
- `"Ana"` é o valor dessa chave
- `"idade"` é outra chave
- `19` é o valor dela


````python
aluno = {
    "nome": "Ana",
    "idade": 19,
    "curso": "Engenharia Civil",
    "altura": 1.80
}

print(aluno)
````

# Acessando valores

Para pegar uma informação específica, usamos a chave entre colchetes.


````python
aluno = {
    "nome": "Ana",
    "idade": 19,
    "curso": "Engenharia Civil"
}

print(aluno["nome"])
print(aluno["idade"])
print(aluno["curso"])
````

# Entendendo o acesso passo a passo

Quando fazemos:

```python
print(aluno["nome"])
```

o Python procura a chave `"nome"` dentro do dicionário e devolve o valor associado a ela.


# Alterando um valor

Dicionários podem ser modificados.


````python
aluno = {
    "nome": "Ana",
    "idade": 19,
    "curso": "Engenharia Civil"
}

print("Antes:", aluno)

aluno["idade"] = 20

print("Depois:", aluno)
````

# Adicionando uma nova informação

Se a chave ainda não existir, ela será criada.


````python
aluno = {
    "nome": "Ana",
    "idade": 19
}

aluno["cidade"] = "Jales"

print(aluno["nome"])
print(aluno["idade"])
print(aluno["cidade"])
````

# Removendo uma informação

Podemos apagar uma chave com `del`.


````python
aluno = {
    "nome": "Ana",
    "idade": 19,
    "cidade": "Jales"
}

print("Antes:", aluno)

del aluno["cidade"]

print("Depois:", aluno)
````

# Exemplo bem simples: estoque de frutas

Um dicionário pode ser usado para guardar quantidades.


````python
frutas = {
    "maçã": 10,
    "banana": 5,
    "laranja": 8
}

print("Quantidade de maçãs:", frutas["maçã"])
print("Quantidade de bananas:", frutas["banana"])
````

# Atualizando quantidades

Imagine que chegaram mais bananas no estoque.


````python
frutas = {
    "maçã": 10,
    "banana": 5,
    "laranja": 8
}

frutas["banana"] += 3

print(frutas)
````

# Percorrendo um dicionário com `for`

Às vezes queremos mostrar tudo o que existe no dicionário.


````python
frutas = {
    "maçã": 10,
    "banana": 8,
    "laranja": 8
}

for chave in frutas:
    print("Chave:", chave, "- Valor:", frutas[chave])
````

# Usando `items()`

O método `items()` devolve cada par **chave e valor** ao mesmo tempo.


````python
frutas = {
    "maçã": 10,
    "banana": 8,
    "laranja": 8
}

for nome, quantidade in frutas.items():
    print(nome, "->", quantidade)
````

# Verificando se uma chave existe

Antes de acessar uma chave, podemos verificar se ela está no dicionário.


````python
aluno = {
    "nome": "Carlos",
    "idade": 18
}
if "nome" in aluno:
    print("A chave nome existe.")
    print("Valor:", aluno["nome"])
if "telefone" in aluno:
    print(aluno["telefone"])
else:
    print("A chave telefone não existe.")
````

# Usando `get()`

O método `get()` é útil porque evita erro quando a chave não existe.


````python
aluno = {
    "nome": "Carlos",
    "idade": 18
}

print(aluno.get("nome"))
print(aluno.get("telefone"))
print(aluno.get("telefone", "Não cadastrado"))
````

# Exemplo com notas de um aluno

Aqui cada chave representa uma disciplina.


````python
notas = {
    "matemática": 7.5,
    "física": 8.0,
    "química": 6.5
}

print("Nota em matemática:", notas["matemática"])
print("Nota em física:", notas["física"])
````

# Calculando média com dicionário

Podemos percorrer os valores e fazer contas com eles.


````python
notas = {
    "matemática": 7.5,
    "física": 8.0,
    "química": 6.5
}

soma = 0

for valor in notas.values():
    soma += valor

media = soma / len(notas)

print("Média:", media)
````

# Dicionário dentro de dicionário

Esse é um passo a mais, mas vamos fazer de forma simples.

Cada aluno tem outro dicionário com suas informações.


````python
turma = {
    "Ana": {"idade": 19, "cidade": "Jales"},
    "Bruno": {"idade": 20, "cidade": "Santa Fé do Sul"}
}

print(turma["Ana"])
print(turma["Ana"]["idade"])
print(turma["Bruno"]["cidade"])
````

# Mini desafio resolvido

Criar um dicionário com 3 produtos e seus preços.
Depois:
- mostrar o preço de um produto
- alterar o preço de outro
- mostrar tudo


````python
produtos = {
    "caderno": 15.0,
    "caneta": 3.5,
    "borracha": 2.0
}

print("Preço do caderno:", produtos["caderno"])
print("Preço do caneta:", produtos["caneta"])

produtos["caneta"] = 4.0

for nome, preco in produtos.items():
    print(nome, "->", preco)
````

# Erros comuns de quem está começando

>Erro 1: esquecer aspas na chave texto

Errado:
```python
aluno = {nome: "Ana"}
```

Certo:
```python
aluno = {"nome": "Ana"}
```

>Erro 2: acessar chave que não existe

```python
print(aluno["telefone"])
```

Se `"telefone"` não existir, dá erro.

>Erro 3: confundir dicionário com lista

Lista usa índices:
```python
nomes[0]
```

Dicionário usa chaves:
```python
aluno["nome"]
```
