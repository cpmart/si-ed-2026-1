Em Python, a forma mais comum de trabalhar com listas lineares é usando `list`.

Exemplo:


````python
alunos = ["Ana", "Bruno", "Carlos", "Daniela"]
print(alunos)
````

# Operações básicas em uma lista Python

Vamos revisar as principais operações, mas agora com atenção ao que acontece na estrutura.


````python
lista = [10, 20, 30]
print("Lista inicial:", lista)

lista.append(40)
print("Depois de append(40):", lista)

lista.insert(1, 15)
print("Depois de insert(1, 15):", lista)

lista.remove(20)
print("Depois de remove(20):", lista)

valor = lista[2]
print("Elemento na posição 2:", valor)
````

## Leitura didática do código

- `append(40)` adiciona o valor no final da lista.
- `insert(1, 15)` insere o valor na posição 1.
- `remove(20)` procura e remove o valor 20.
- `lista[2]` acessa diretamente o elemento da posição 2.

O ponto mais importante é perceber que algumas operações são simples e outras exigem deslocamento ou busca.


# O que acontece por trás das operações?

A lista do Python permite acesso direto por posição.

```text
lista = [10, 20, 30, 40]

posição:   0    1    2    3
valor:    10   20   30   40
```

Se acessarmos `lista[2]`, o Python vai diretamente ao elemento da posição 2.

Mas ao inserir no início, os elementos precisam abrir espaço.

```text

Inserir 5 na posição 0:
[5, 10, 20, 30, 40]

O 10, 20, 30 e 40 precisaram ser deslocados.
```


````python
lista = [10, 20, 30, 40]
print("Antes:", lista)

lista.insert(0, 5)
print("Depois de inserir no início:", lista)
````

# Relação com complexidade de algoritmos

Nem toda operação em lista tem o mesmo custo.

Resumo em texto:

- Acessar por posição em uma lista Python é rápido: `O(1)`.
- Inserir no final costuma ser rápido: aproximadamente `O(1)`.
- Inserir no início exige deslocar elementos: `O(n)`.
- Buscar um valor exige percorrer a lista: `O(n)`.
- Remover um valor também pode exigir busca e reorganização: `O(n)`.

Essa análise ajuda a entender por que existem outras formas de organizar dados.


````python
# Exemplo simples de busca linear em uma list Python
numeros = [8, 3, 10, 25, 7]
procurado = 25

for posicao, valor in enumerate(numeros):
    print(f"Verificando posição {posicao}: valor {valor}")
    if valor == procurado:
        print(f"Valor encontrado na posição {posicao}")
        break
````

# Aplicação de lista linear: pilha

Além de representar uma sequência simples, uma lista linear também pode ser usada para implementar outras estruturas de dados.

Uma das estruturas mais importantes é a **pilha**.

A pilha segue a regra **LIFO**:

```text
Last In, First Out
Último a entrar, primeiro a sair
```

Exemplo do cotidiano:

```text
Pilha de pratos

[ prato 3 ]  <- topo
[ prato 2 ]
[ prato 1 ]
```

Quando colocamos um novo prato, ele entra no topo.
Quando retiramos um prato, retiramos primeiro o que está no topo.

## Pilha usando `list` em Python

Como já conhecem `list`, podemos implementar uma pilha de forma simples.

Usaremos:

- `append()` para empilhar um elemento;
- `pop()` para desempilhar o último elemento inserido.

```text
pilha = []

empilhar 10 -> [10]
empilhar 20 -> [10, 20]
empilhar 30 -> [10, 20, 30]

desempilhar -> remove 30
```

````python
pilha = []

print("Pilha inicial:", pilha)

# Empilhando elementos
pilha.append(10)
print("Depois de empilhar 10:", pilha)

pilha.append(20)
print("Depois de empilhar 20:", pilha)

pilha.append(30)
print("Depois de empilhar 30:", pilha)

# Desempilhando o último elemento
removido = pilha.pop()
print("Elemento desempilhado:", removido)
print("Pilha após desempilhar:", pilha)
````

## Leitura didática do exemplo

Na pilha, não removemos qualquer elemento.

A regra é sempre mexer no **topo** da pilha.

No exemplo anterior:

```text
Pilha antes do pop(): [10, 20, 30]
Topo: 30

Após pop(): [10, 20]
```

Portanto, o valor `30` foi removido porque foi o último valor inserido.

## Criando uma classe Pilha

Para aproximar esse conteúdo da ideia de estrutura de dados, podemos encapsular a pilha em uma classe.

Isso ajuda o aluno a perceber que uma estrutura de dados não é apenas uma lista solta, mas um conjunto de dados com regras de uso.

````python
class Pilha:
    def __init__(self):
        self.itens = []

    def esta_vazia(self):
        return len(self.itens) == 0

    def empilhar(self, valor):
        self.itens.append(valor)
        print(f"Empilhei {valor}:", self.itens)

    def desempilhar(self):
        if self.esta_vazia():
            print("A pilha está vazia. Não há elemento para remover.")
            return None

        valor_removido = self.itens.pop()
        print(f"Desempilhei {valor_removido}:", self.itens)
        return valor_removido

    def topo(self):
        if self.esta_vazia():
            print("A pilha está vazia. Não existe topo.")
            return None

        return self.itens[-1]

    def exibir(self):
        print("Pilha:", self.itens)


pilha = Pilha()
pilha.empilhar("A")
pilha.empilhar("B")
pilha.empilhar("C")

print("Topo atual:", pilha.topo())

pilha.desempilhar()
pilha.exibir()
````

## Relação entre lista e pilha

A pilha pode ser implementada usando uma lista Python, mas ela não deve ser usada como uma lista comum.

Em uma lista comum, podemos inserir, remover ou acessar elementos em várias posições.

Em uma pilha, o acesso é controlado por uma regra:

```text
Só podemos inserir e remover pelo topo.
```

Isso mostra que a diferença não está apenas no código, mas na **forma como a estrutura é utilizada**.

# Por que estudar lista encadeada?

A lista encadeada é uma alternativa de organização dos dados.

Em vez de guardar os elementos lado a lado, cada elemento guarda:

- um valor;
- uma referência para o próximo elemento.

Visualmente:

```text
[10 | próximo] -> [20 | próximo] -> [30 | None]
```

A lista encadeada não substitui a `list` do Python em todos os casos. Ela é estudada porque ajuda a entender:

- organização de dados na memória;
- estruturas dinâmicas;
- referências entre objetos;
- custo das operações;
- base para pilhas, filas, árvores e grafos.


## Ideia central

Em uma lista Python, usamos índice:

```python
lista[2]
```

Em uma lista encadeada, não existe acesso direto por índice.

Para chegar ao terceiro elemento, é preciso começar no primeiro e seguir os próximos nós.

```text
início -> 10 -> 20 -> 30 -> None
```


# Criando nós com classe

Usando `class`, podemos usar esse conhecimento para criar um nó.

Um nó é um objeto que guarda duas informações:

1. `valor`: o dado armazenado;
2. `proximo`: a referência para o próximo nó.


````python
class No:
    def __init__(self, valor):
        self.valor = valor
        self.proximo = None

# Criando um nó isolado
no1 = No(10)

print("Valor do nó:", no1.valor)
print("Próximo nó:", no1.proximo)
````

Neste momento ainda não temos uma lista encadeada.

Temos apenas um objeto do tipo `No`.

```text
[10 | None]
```


# Ligando objetos: a ideia de referência

Agora vamos criar três nós e ligar um ao outro.

A palavra importante aqui é `referência`.

Quando fazemos `no1.proximo = no2`, estamos dizendo que o primeiro nó aponta para o segundo.


````python
no1 = No(10)
no2 = No(20)
no3 = No(30)

no1.proximo = no2
no2.proximo = no3

print("no1 guarda:", no1.valor)
print("Depois do no1 vem:", no1.proximo.valor)
print("Depois do no2 vem:", no2.proximo.valor)
print("Depois do no3 vem:", no3.proximo)
````

Representação visual:

```text
no1              no2              no3
[10 |  * ]  ->   [20 |  * ]  ->   [30 | None]
```

**Importante:**

> Não é algo mágico. O atributo `proximo` apenas guarda uma referência para outro objeto.


# Percorrendo uma lista encadeada

Para percorrer uma lista encadeada, usamos uma variável auxiliar.

Normalmente chamamos essa variável de `atual`.

Ela começa no primeiro nó e vai caminhando até chegar em `None`.


````python
atual = no1

while atual is not None:
    print("Estou no nó com valor:", atual.valor)
    atual = atual.proximo
````

## Rastreamento do algoritmo

Para a lista:

```text
10 -> 20 -> 30 -> None
```

O algoritmo faz:

```text
atual = 10
imprime 10
atual = próximo de 10, que é 20
imprime 20
atual = próximo de 20, que é 30
imprime 30
atual = próximo de 30, que é None
para o while
```

Essa etapa é essencial. Entendendo esse percurso, é possível entender inserção, busca e remoção.


# Criando a classe ListaEncadeada passo a passo

Agora vamos encapsular o funcionamento dentro de uma classe.

A lista encadeada precisa guardar uma referência para o primeiro nó.

Chamaremos essa referência de `inicio`.


````python
class ListaEncadeada:
    def __init__(self):
        self.inicio = None

    def esta_vazia(self):
        return self.inicio is None

# Criando uma lista vazia
lista_enc = ListaEncadeada()
print("A lista está vazia?", lista_enc.esta_vazia())
````

Visualmente, uma lista vazia é:

```text
inicio -> None
```

Quando adicionarmos um primeiro nó, teremos:

```text
inicio -> [10 | None]
```


# Inserção no início

Inserir no início é uma das operações mais importantes para entender lista encadeada.

A lógica é:

1. Criar um novo nó.
2. Fazer o novo nó apontar para o antigo início.
3. Atualizar o início da lista para o novo nó.

Exemplo:

```text
Antes:
inicio -> [20] -> [30] -> None

Inserindo 10 no início:
novo -> [10]
novo.proximo aponta para [20]
inicio passa a apontar para [10]

Depois:
inicio -> [10] -> [20] -> [30] -> None
```


````python
class ListaEncadeada:
    def __init__(self):
        self.inicio = None

    def esta_vazia(self):
        return self.inicio is None

    def inserir_inicio(self, valor):
        novo_no = No(valor)
        novo_no.proximo = self.inicio
        self.inicio = novo_no

    def exibir(self):
        atual = self.inicio
        while atual is not None:
            print(atual.valor, end=" -> ")
            atual = atual.proximo
        print("None")

lista_enc = ListaEncadeada()
lista_enc.inserir_inicio(30)
lista_enc.inserir_inicio(20)
lista_enc.inserir_inicio(10)

lista_enc.exibir()
````

## Observação didática

A inserção no início não precisa percorrer a lista.

Por isso, em uma lista encadeada simples, inserir no início tem custo `O(1)`.


# Inserção no fim

Para inserir no fim, precisamos caminhar até o último nó.

O último nó é aquele cujo `proximo` vale `None`.

Exemplo:

```text
10 -> 20 -> 30 -> None
```

Para inserir 40:

```text
1. começa no 10
2. vai para o 20
3. vai para o 30
4. percebe que 30.proximo é None
5. faz 30.proximo apontar para 40
```


````python
class ListaEncadeada:
    def __init__(self):
        self.inicio = None

    def esta_vazia(self):
        return self.inicio is None

    def inserir_inicio(self, valor):
        novo_no = No(valor)
        novo_no.proximo = self.inicio
        self.inicio = novo_no

    def inserir_fim(self, valor):
        novo_no = No(valor)

        if self.inicio is None:
            self.inicio = novo_no
            return

        atual = self.inicio
        while atual.proximo is not None:
            print("Ainda não é o último. Estou no valor:", atual.valor)
            atual = atual.proximo

        print("Cheguei no último nó:", atual.valor)
        atual.proximo = novo_no

    def exibir(self):
        atual = self.inicio
        while atual is not None:
            print(atual.valor, end=" -> ")
            atual = atual.proximo
        print("None")

lista_enc = ListaEncadeada()
lista_enc.inserir_fim(10)
lista_enc.inserir_fim(20)
lista_enc.inserir_fim(30)
lista_enc.inserir_fim(40)

lista_enc.exibir()
````

## Observação didática

Na implementação acima, inserir no fim exige percorrer a lista até o último nó.

Por isso, o custo é `O(n)`.


# Evolução: guardando também o último nó

Na implementação anterior, a lista guardava apenas uma referência para o primeiro nó, chamada `inicio`.

Isso funciona, mas existe um detalhe importante: para inserir um novo valor no final, o algoritmo precisa começar no primeiro nó e caminhar até encontrar o último.

Visualmente:

```text
inicio
  ↓
[10 | próximo] → [20 | próximo] → [30 | None]
```

Para inserir o valor `40`, precisamos percorrer:

```text
10 → 20 → 30 → inserir 40
```

Ou seja, inserir no final custa `O(n)`, porque depende da quantidade de nós da lista.

Uma melhoria possível é guardar também uma referência para o último nó da lista, chamada aqui de `fim`.

```text
inicio                              fim
  ↓                                  ↓
[10 | próximo] → [20 | próximo] → [30 | None]
```

Assim, para inserir no final, não precisamos mais percorrer toda a lista. Basta ligar o último nó ao novo nó e atualizar a referência `fim`.

## Ideia da melhoria

Agora a lista encadeada terá duas referências principais:

- `inicio`: aponta para o primeiro nó da lista.
- `fim`: aponta para o último nó da lista.

Quando a lista estiver vazia:

```text
inicio → None
fim    → None
```

Quando inserirmos o primeiro valor, o mesmo nó será, ao mesmo tempo, o início e o fim da lista:

```text
inicio
  ↓
[10 | None]
  ↑
fim
```

Quando inserirmos novos valores no final, apenas atualizamos a ligação do antigo último nó e depois movemos o `fim` para o novo nó.

````python
class No:
    def __init__(self, valor):
        self.valor = valor
        self.proximo = None


class ListaEncadeadaComFim:
    def __init__(self):
        # No início, a lista está vazia.
        # Portanto, não existe primeiro nó nem último nó.
        self.inicio = None
        self.fim = None

    def esta_vazia(self):
        return self.inicio is None

    def inserir_inicio(self, valor):
        novo_no = No(valor)

        # Caso 1: lista vazia
        if self.esta_vazia():
            self.inicio = novo_no
            self.fim = novo_no
        else:
            # Caso 2: lista já possui elementos
            novo_no.proximo = self.inicio
            self.inicio = novo_no

    def inserir_fim(self, valor):
        novo_no = No(valor)

        # Caso 1: lista vazia
        # O novo nó será o primeiro e também o último.
        if self.esta_vazia():
            self.inicio = novo_no
            self.fim = novo_no
        else:
            # Caso 2: lista já possui elementos
            # O antigo último nó passa a apontar para o novo nó.
            self.fim.proximo = novo_no

            # Agora o novo nó passa a ser o último.
            self.fim = novo_no

    def exibir(self):
        atual = self.inicio

        while atual is not None:
            print(atual.valor, end=" -> ")
            atual = atual.proximo

        print("None")

    def mostrar_inicio_e_fim(self):
        if self.esta_vazia():
            print("Lista vazia")
        else:
            print("Início:", self.inicio.valor)
            print("Fim:", self.fim.valor)
````

## Testando a lista com referência para o fim

Agora vamos inserir elementos no final da lista.

A diferença principal é que o método `inserir_fim()` não precisa mais usar `while` para procurar o último nó.

Ele já sabe quem é o último, porque a lista guarda essa informação em `self.fim`.


````python
lista = ListaEncadeadaComFim()

lista.inserir_fim(10)
lista.inserir_fim(20)
lista.inserir_fim(30)

lista.exibir()
lista.mostrar_inicio_e_fim()
````

## Rastreamento passo a passo da inserção no fim

Vamos acompanhar a lógica de uma inserção no final usando a referência `fim`.

Situação inicial:

```text
inicio                              fim
  ↓                                  ↓
[10 | próximo] → [20 | próximo] → [30 | None]
```

Queremos inserir `40`.

### Passo 1: criar o novo nó

```text
[40 | None]
```

### Passo 2: ligar o antigo último nó ao novo nó

```text
[30 | próximo] → [40 | None]
```

### Passo 3: atualizar o `fim`

```text
inicio                                           fim
  ↓                                               ↓
[10 | próximo] → [20 | próximo] → [30 | próximo] → [40 | None]
```

Observe que não foi necessário caminhar desde o início da lista.

````python
lista.inserir_fim(40)

lista.exibir()
lista.mostrar_inicio_e_fim()
````

# Busca em lista encadeada

A busca em uma lista encadeada é feita percorrendo nó por nó.

Não conseguimos acessar diretamente uma posição como fazemos com `lista[2]`.


````python
class ListaEncadeada:
    def __init__(self):
        self.inicio = None

    def inserir_fim(self, valor):
        novo_no = No(valor)

        if self.inicio is None:
            self.inicio = novo_no
            return

        atual = self.inicio
        while atual.proximo is not None:
            atual = atual.proximo
        atual.proximo = novo_no

    def buscar(self, valor_procurado):
        atual = self.inicio
        posicao = 0

        while atual is not None:
            print(f"Verificando posição {posicao}: valor {atual.valor}")

            if atual.valor == valor_procurado:
                return posicao

            atual = atual.proximo
            posicao += 1

        return -1

    def exibir(self):
        atual = self.inicio
        while atual is not None:
            print(atual.valor, end=" -> ")
            atual = atual.proximo
        print("None")

lista_enc = ListaEncadeada()
lista_enc.inserir_fim(8)
lista_enc.inserir_fim(3)
lista_enc.inserir_fim(10)
lista_enc.inserir_fim(25)
lista_enc.inserir_fim(7)

lista_enc.exibir()
resultado = lista_enc.buscar(25)
print("Resultado da busca:", resultado)
````

Se o valor for encontrado, o método retorna a posição lógica em que ele apareceu.

Se não for encontrado, retorna `-1`.


# Remoção no início

Remover o primeiro nó é simples.

Basta fazer o início apontar para o segundo nó.

```text
Antes:
inicio -> [10] -> [20] -> [30] -> None

Depois de remover o início:
inicio -> [20] -> [30] -> None
```


````python
class ListaEncadeada:
    def __init__(self):
        self.inicio = None

    def inserir_fim(self, valor):
        novo_no = No(valor)

        if self.inicio is None:
            self.inicio = novo_no
            return

        atual = self.inicio
        while atual.proximo is not None:
            atual = atual.proximo
        atual.proximo = novo_no

    def remover_inicio(self):
        if self.inicio is None:
            print("A lista está vazia. Nada para remover.")
            return None

        valor_removido = self.inicio.valor
        self.inicio = self.inicio.proximo
        return valor_removido

    def exibir(self):
        atual = self.inicio
        while atual is not None:
            print(atual.valor, end=" -> ")
            atual = atual.proximo
        print("None")

lista_enc = ListaEncadeada()
lista_enc.inserir_fim(10)
lista_enc.inserir_fim(20)
lista_enc.inserir_fim(30)

print("Antes:")
lista_enc.exibir()

removido = lista_enc.remover_inicio()
print("Valor removido:", removido)

print("Depois:")
lista_enc.exibir()
````

# Remoção por valor

Remover um valor no meio da lista exige cuidado.

Precisamos guardar duas referências:

- `atual`: nó que está sendo analisado;
- `anterior`: nó que vem antes do atual.

Exemplo:

```text
10 -> 20 -> 30 -> 40 -> None
```

Para remover o 30:

```text
anterior = 20
atual = 30

Fazemos:
20.proximo = 40
```

Assim, o 30 deixa de fazer parte da sequência.


````python
class ListaEncadeada:
    def __init__(self):
        self.inicio = None

    def inserir_fim(self, valor):
        novo_no = No(valor)

        if self.inicio is None:
            self.inicio = novo_no
            return

        atual = self.inicio
        while atual.proximo is not None:
            atual = atual.proximo
        atual.proximo = novo_no

    def remover_valor(self, valor):
        if self.inicio is None:
            return False

        if self.inicio.valor == valor:
            self.inicio = self.inicio.proximo
            return True

        anterior = self.inicio
        atual = self.inicio.proximo

        while atual is not None:
            print("Anterior:", anterior.valor, "| Atual:", atual.valor)

            if atual.valor == valor:
                anterior.proximo = atual.proximo
                return True

            anterior = atual
            atual = atual.proximo

        return False

    def exibir(self):
        atual = self.inicio
        while atual is not None:
            print(atual.valor, end=" -> ")
            atual = atual.proximo
        print("None")

lista_enc = ListaEncadeada()
for numero in [10, 20, 30, 40]:
    lista_enc.inserir_fim(numero)

print("Antes:")
lista_enc.exibir()

removido = lista_enc.remover_valor(30)
print("Removeu?", removido)

print("Depois:")
lista_enc.exibir()
````

# Comparação final: list x lista encadeada

A `list` do Python é excelente para muitos casos práticos, especialmente quando precisamos acessar elementos por posição.

A lista encadeada é importante para entender estruturas dinâmicas e referências entre objetos.

Comparação em texto:

## Lista Python

- Permite acesso direto por índice.
- É simples de usar.
- Inserções no início ou no meio podem exigir deslocamento.
- É a estrutura mais indicada para muitos problemas comuns em Python.

## Lista Encadeada

- Não permite acesso direto por índice.
- Cada nó aponta para o próximo.
- Inserir no início é rápido.
- Percorrer e buscar exigem caminhar nó por nó.
- É importante como base para outras estruturas de dados.

## Complexidade aproximada

- Acesso por posição em `list`: `O(1)`.
- Acesso por posição em lista encadeada: `O(n)`.
- Inserção no início em `list`: `O(n)`.
- Inserção no início em lista encadeada: `O(1)`.
- Busca nas duas estruturas: `O(n)`.


# Outras variações de listas encadeadas

Até aqui, estudamos a **lista encadeada simples**.

Nela, cada nó conhece apenas o próximo nó:

```text
[10 | próximo] -> [20 | próximo] -> [30 | None]
```

Existem outras variações importantes. Nesta aula, vamos apenas entender a ideia geral, sem implementar o código completo.

## Lista encadeada circular

Na lista encadeada circular, o último nó não aponta para `None`.

Ele aponta novamente para o primeiro nó.

```text
início
  |
  v
[10] -> [20] -> [30]
 ^               |
 |_______________|
```

A principal diferença é que não existe um fim marcado por `None`.

Isso pode ser útil quando precisamos percorrer elementos de forma contínua, como em:

- sistemas de rodízio;
- escalas circulares;
- alternância entre jogadores;
- listas de tarefas repetitivas.

### Exemplo conceitual de lista circular

Imagine um sistema que alterna a vez entre três alunos:

```text
Ana -> Bruno -> Carlos -> Ana -> Bruno -> Carlos -> ...
```

Quando chega em Carlos, o próximo volta a ser Ana.

Essa é a ideia da lista encadeada circular.

O cuidado principal é evitar laços infinitos, pois a lista não termina em `None`.

## Lista duplamente encadeada

Na lista duplamente encadeada, cada nó guarda duas referências:

- uma para o nó anterior;
- outra para o próximo nó.

```text
None <- [10] <-> [20] <-> [30] -> None
```

Cada nó pode ser representado mentalmente assim:

```text
[ anterior | valor | próximo ]
```

Isso permite caminhar nos dois sentidos:

```text
início -> fim
fim -> início
```

### Exemplo conceitual de lista duplamente encadeada

Um exemplo fácil de entender é uma playlist de músicas.

```text
Música 1 <-> Música 2 <-> Música 3
```

O usuário pode:

- avançar para a próxima música;
- voltar para a música anterior.

Em uma lista simplesmente encadeada, voltar seria mais difícil, pois cada nó só conhece o próximo.

## Lista duplamente encadeada circular

A lista duplamente encadeada circular combina as duas ideias anteriores.

Ela é:

- duplamente encadeada, porque cada nó aponta para o anterior e para o próximo;
- circular, porque o último nó aponta para o primeiro e o primeiro aponta para o último.

```text
 ______________________
|                     v
[10] <-> [20] <-> [30]
 ^                    |
 |____________________|
```

Em uma representação mais completa:

```text
[10] <-> [20] <-> [30]
  ^                 |
  |_________________|
```

Também é possível voltar do primeiro para o último.

### Exemplo conceitual de lista duplamente encadeada circular

Um exemplo é um menu circular de opções.

```text
Opção 1 <-> Opção 2 <-> Opção 3 <-> Opção 1
```

O usuário pode navegar:

- para frente;
- para trás;
- e continuar circulando pelas opções.

Esse tipo de estrutura aparece em situações em que a navegação precisa ser contínua nos dois sentidos.

## Comparação conceitual entre as variações

Resumo em texto:

- Lista encadeada simples: cada nó aponta apenas para o próximo.
- Lista encadeada circular: o último nó aponta novamente para o primeiro.
- Lista duplamente encadeada: cada nó aponta para o anterior e para o próximo.
- Lista duplamente encadeada circular: permite navegação nos dois sentidos e o percurso é circular.

A escolha entre essas estruturas depende do problema que queremos resolver.

Quanto mais referências cada nó guarda, mais flexível fica a navegação. Porém, a implementação também fica mais trabalhosa, porque cada inserção e remoção precisa atualizar mais ligações.
