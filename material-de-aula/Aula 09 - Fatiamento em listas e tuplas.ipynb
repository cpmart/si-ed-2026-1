# Fatiamento em Listas e Tuplas

O **fatiamento** em Python é uma técnica usada para acessar partes específicas de sequências como listas, strings e tuplas. O fatiamento segue a sintaxe:

```python
sequencia[início:fim:passo]
```

Onde:
- `inicio`: índice de onde o fatiamento começa (inclusivo).
- `fim`: índice onde o fatiamento termina (exclusivo).
- `passo`: a quantidade de posições que a fatia irá pular (opcional).

## Exemplos de Fatiamento com Listas

### Fatiar os primeiros 3 elementos

````python
lista = [0, 1, 2, 3, 4, 5]
print(lista[:3])  

# saída:
````

### Fatiar do índice 2 até o final

````python
lista = [0, 1, 2, 3, 4, 5, 20, 30]
print(lista[2:])  # Do índice 2 até o final

# saída: 
````

### Fatiar os elementos entre os índices 1 e 4

````python
lista = [0, 1, 2, 3, 4, 5]
print(lista[1:4])  # Do índice 1 até o 3 (o índice 4 é excluído)

# saída:
````

### Fatiar com passo (intervalo)

````python
lista = [0, 1, 2, 3, 4, 5, 6, 7, 8]
print(lista[2::3])  # Do início ao fim, pulando de 2 em 2 elementos

# saída:
````

### Fatiar a lista ao contrário

````python
lista = [0, 1, 2, 3, 4, 5]
print(lista[::-1])  # Todos os elementos, mas na ordem reversa

# saída:
````

### Fatiar com passo negativo, pegando elementos alternados ao contrário

````python
lista = [0, 1, 2, 3, 4, 5, 6]
print(lista[6:1:-2])  # Do índice 5 ao índice 2, pulando de 2 em 2 (invertido)

# saída: 
````

### Fatiar uma lista ignorando o primeiro e o último elementos

````python
lista = [10, 123, 21, 32, 45, 57]
print(lista[1:-1])  # Todos os elementos menos o primeiro e o último

# saída:
````

## Exemplos de Fatiamento com Strings

### Fatiar os primeiros 4 caracteres de uma string

````python
texto = "Python"
print(texto[:4])  # Do início até o índice 3 (o índice 4 é excluído)

# saída:
````

### Fatiar os últimos 3 caracteres de uma string

````python
texto = "Python"
print(texto[-3:])  # Do terceiro último caractere até o final

# saída:
````

### Fatiar e reverter a string

````python
texto = "Python"
print(texto[::-1])  # Todos os caracteres, mas na ordem inversa

# saída: 
````

## Exemplos de Fatiamento com Tuplas

### Fatiar uma tupla

````python
tupla = (10, 20, 30, 40, 50)
print(tupla[1:4])  # Do índice 1 até o 3 (o índice 4 é excluído)

# saída:
````

### Fatiar uma tupla de forma reversa

````python
tupla = (10, 20, 30, 40, 50)
print(tupla[::-1])  # Todos os elementos, mas na ordem inversa

# saída:
````
