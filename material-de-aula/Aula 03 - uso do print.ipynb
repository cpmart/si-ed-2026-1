# A função ```print()```

O comando print em Python é utilizado para exibir dados na tela. Existem várias formas de utilizar o print, e a função é bastante versátil. Abaixo, seguem as principais formas de utilizá-lo, com exemplos de código para cada caso.

## 1. **print() Básico**
Imprime uma string ou números diretamente.

```(python)
print("Hello, World!")
print(42)
```

**Saída:**
```
Hello, World!
42
```

## 2. **Print com Variáveis**
Imprime o valor de variáveis.

```
name = "Alice"
age = 30
print(name)
print(age)
```

**Saída:**
```
Alice
30
```

## 3. **Print com Formatação de Strings**
Utiliza `format()` e f-strings para inserir variáveis em uma string. 

O `format()` substituiu o `%`.  Antigamente era usado:

```
nome = "João""
idade = 20
grana = 50.00
"{} tem {} anos e R${} no bolso." % (nome, idade, grana)
```
**Saída:**
```
João tem 20 anos e R$50.00 no bolso.
```

Agora é usado assim: 

```
name = "Alice"
age = 30
print("Name: {}, Age: {}".format(name, age))
print(f"Name: {name}, Age: {age}")
```

**Saída:**
```
Name: Alice, Age: 30
Name: Alice, Age: 30
```

## 4. **Print com Argumentos Posicionais e Nomeados**
Formatação com argumentos posicionais e nomeados em `format()`.

```
print("{0} {1}".format("Hello", "World"))
print("{greeting} {name}".format(greeting="Hello", name="Alice"))
```

**Saída:**
```
Hello World
Hello Alice
```

## 5. **Print com Múltiplos Argumentos**
Imprime vários valores de uma vez, separados por espaço por padrão.

```
print("Hello", "World", 123)
```

**Saída:**
```
Hello World 123
```

## 6. **Print com `sep` e `end`**
Controla o separador e o caractere de finalização.

```
print("Hello", "World", sep="-")
print("Hello", end=", ")
print("World")
```

**Saída:**
```
Hello-World
Hello, World
```

## 7. **Print em Múltiplas Linhas**
Usa `\n` para quebrar linha ou string multilinha com `"""`.

```

print("""This is
a multiline
string""")
```

**Saída:**
```

World

a multiline
string
```

## 8. **Print de Listas e Dicionários**
Imprime listas e dicionários diretamente.


**Saída:**

## 10. **Print com Concatenando Strings**
Usa o operador `+` para concatenar strings.

```

```

**Saída:**
```

```

## 11. **Print de Objetos Customizados**
Imprime representações de objetos customizados usando `__str__` e `__repr__`.

````python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def __str__(self):
        return f"Person(name={self.name}, age={self.age})"
    
    def dobro(self):
        return self.age*2

person = Person("Alice", 30)
pessoa2 = Person("Cristiano", 25)
print(pessoa2.name)
print(pessoa2.age)
print(pessoa2.dobro())
````


**Saída:**
```
Person(name=Alice, age=30)
```

## 12. **Print com `pprint` para Dicionários**
Exibe dicionários de forma mais legível com o módulo `pprint`.


**Saída:**
```

 'city': 'New York',
```

## 13. **Print com `join` para Concatenar Sequências**
Concatena elementos de uma lista ou tupla em uma string.


````python
fruits = ["apple", "banana", "cherry"]
print(", ".join(fruits))

print("apple", "banana", "cherry", sep=", ")
print(*fruits, sep=", ")
#**Saída:**
# apple, banana, cherry
````


## 14. **Print de Lambda Expressions**

## Formato f-strings

O formato `print(f"")`, conhecido como **f-strings**, é amplamente utilizado em Python para formatação de strings de maneira eficiente e legível. As f-strings foram introduzidas no Python 3.6 e permitem a interpolação de variáveis diretamente em uma string literal, usando a sintaxe `f"{}"`.

Mais Exemplos de como usar f-strings com a função `print`:

### 1. **Incorporar Variáveis em uma String**

```
name = "Alice"
age = 30
print(f"Name: {name}, Age: {age}")
```

**Saída:**
```
Name: Alice, Age: 30
```

### 2. **Formatar Números com Casas Decimais**

```
pi = 3.14159265358979
print(f"Value of pi to 2 decimal places: {pi:.2f}")
```

**Saída:**
```
Value of pi to 2 decimal places: 3.14
```

### 3. **Inserir Resultados de Expressões**

```
a = 5
b = 10
print(f"The sum of {a} and {b} is {a + b}")
```

**Saída:**
```
The sum of 5 and 10 is 15
```

### 4. **Alinhar e Preencher Textos**

```
text = "Python"
print(f"{text:<10} | {text:^10} | {text:>10}")
```

**Saída:**
```
Python    |   Python   |     Python
```

### 5. **Usar Funções Dentro das f-strings**

```
import math
number = 16
print(f"The square root of {number} is {math.sqrt(number):.2f}")
```

**Saída:**
```
The square root of 16 is 4.00
```

As f-strings são preferidas por sua clareza e eficiência em relação a outras formas de formatação, como `format()` e `%`.


O operador `%` para formatação de strings é uma **maneira antiga**, mas ainda utilizada, para formatar strings em Python. Aqui estão alguns exemplos de como usar `%` com a função `print`:

### 1. **Imprimir Números Inteiros**

```
age = 25
print("Age: %d" % age)
```

**Saída:**
```
Age: 25
```

### 2. **Imprimir Números de Ponto Flutuante**

```
pi = 3.14159
print("Value of pi: %.2f" % pi)
```

**Saída:**
```
Value of pi: 3.14
```

### 3. **Imprimir Strings**

```
name = "Alice"
print("Name: %s" % name)
```

**Saída:**
```
Name: Alice
```

### 4. **Imprimir Múltiplas Variáveis**

```
name = "Bob"
age = 30
height = 175.5
print("Name: %s, Age: %d, Height: %.1f cm" % (name, age, height))
```

**Saída:**
```
Name: Bob, Age: 30, Height: 175.5 cm
```

### 5. **Imprimir Representações Hexadecimal e Octal**

```
number = 255
print("Hexadecimal: %x" % number)
print("Octal: %o" % number)
```

**Saída:**
```
Hexadecimal: ff
Octal: 377
```

### 6. **Imprimir com Preenchimento e Alinhamento**

```
value = 42
print("Value with padding: |%5d|" % value)  # Pad with spaces
print("Value with padding: |%05d|" % value) # Pad with zeros
```

**Saída:**
```
Value with padding: |   42|
Value with padding: |00042|
```

### 7. **Imprimir Valores Percentuais**

```
percentage = 85.567
print("Completion: %.1f%%" % percentage)
```

**Saída:**
```
Completion: 85.6%
```

### 8. **Imprimir Vários Tipos de Dados em uma Única Linha**

```
name = "Charlie"
age = 28
height = 182.3
print("Name: %s, Age: %d, Height: %.2f" % (name, age, height))
```

**Saída:**
```
Name: Charlie, Age: 28, Height: 182.30
```

Esses exemplos mostram como o operador `%` pode ser usado para formatar strings em Python, manipulando diferentes tipos de dados e controlando a forma como são exibidos.
