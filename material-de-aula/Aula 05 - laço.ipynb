# Laço de Repetição

## 1. `while` (Loop/Laço)
O laço while repete um bloco de código enquanto a condição for verdadeira.

### Exemplo 1 - Exemplo básico com contagem

Neste exemplo, o loop while continua até que o valor da variável contador ultrapasse 5.

```python
contador = 1
while contador <= 5:
    print("Contagem:", contador)
    contador += 1
```

### Exemplo 2 - Loop infinito com condição de parada

Neste caso, o loop continua solicitando a senha até que o valor correto seja inserido.

```python
senha = ""
while senha != "1234":
    senha = input("Digite a senha: ")
    if senha != "1234":
        print("Senha incorreta, tente novamente.")
print("Acesso concedido.")
```

### Exemplo 3 - Aqui, o loop é interrompido com o uso de break quando n ultrapassa 5, mesmo sendo um loop while True.

```python
n = 1
while True:
    print(n)
    n += 1
    if n > 5:
        break
print('fim')
```

### Exemplo 3.1 - Uso do continue

```python
n = 1
while True:
    print(n)
    if n > 5:
        continue
    n += 1
```

### Exemplo 4: Contagem regressiva de 5 a 1

```python
contador = 5
while contador > 0:
    print(contador)
    contador -= 1

# ou
contador = 5
while contador: #não teste se é maior que 0
    print(contador)
    contador -= 1
```

## 2. `for` (Loop/Laço)
O laço for é usado para iterar sobre uma sequência (como uma lista, tupla, string ou intervalo de números).

O laço `for` em Python é bastante versátil e pode ser usado de várias formas para iterar sobre sequências e coleções. Abaixo, vou mostrar diferentes maneiras de utilizar o `for` em Python.

#### 1. Repetindo uma ação um número fixo de vezes

```python
for _ in range(0,3):
    print(f"Executando ação")
```

#### 2. Usando for para gerar uma tabuada

```python
numero = int(input("Valor: "))
for i in range(1, 11):
    print(f"{numero} x {i} = {numero * i}")
```

#### 3. Usando `range()`
O `range()` gera uma sequência de números, o que é muito útil para laços `for` que precisam de uma sequência de inteiros.

```python
# De 0 a 4 (5 não é incluído)
for i in range(5):
    print(i)

# De 2 a 9
for i in range(2, 10):
    print(i)

# De 0 a 10, pulando de 2 em 2
for i in range(0, 11, 2):
    print(i)
```

#### 4. Iterando sobre uma Lista
O `for` pode ser usado para iterar sobre todos os elementos de uma lista.

```python
numeros = [1, 2, 3, 4, 5]
for numero in numeros:
    print(numero)
```

#### 5. Iterando sobre uma String
O `for` pode ser usado para iterar sobre cada caractere de uma string.

```python
texto = "Python"
for letra in texto:
    print(letra)
```

#### 6. Iterando sobre um Dicionário
Ao iterar sobre um dicionário, o `for` retorna as chaves por padrão. Você pode acessar as chaves e os valores separadamente.

```python
dicionario = {'a': 1, 'b': 2, 'c': 3}

# Iterando sobre as chaves
for chave in dicionario:
    print(chave)

# Iterando sobre as chaves e valores
for chave, valor in dicionario.items():
    print(f"Chave: {chave}, Valor: {valor}")
```

#### 7. Iterando sobre um Conjunto (Set)
Os conjuntos são coleções não ordenadas de itens únicos. Você pode iterar sobre eles usando `for`.

```python
meu_conjunto = {1, 2, 3, 4, 5}
for item in meu_conjunto:
    print(item)
```

#### 8. Iterando sobre uma Lista com Índices Usando `enumerate()`
Se você precisa do índice e do valor de uma lista, pode usar `enumerate()`.

```python
frutas = ['maçã', 'banana', 'cereja']
for indice, fruta in enumerate(frutas):
    print(f"Índice: {indice}, Fruta: {fruta}")
```

#### 9. Iterando sobre Múltiplas Listas com `zip()`
Você pode iterar sobre duas ou mais listas ao mesmo tempo usando `zip()`.

```python
nomes = ['Alice', 'Bob', 'Charlie']
idades = [25, 30, 35]

for nome, idade in zip(nomes, idades):
    print(f"{nome} tem {idade} anos.")
```

#### 10. Compreensão de Listas (List Comprehension)
A compreensão de listas é uma forma concisa de criar listas em uma única linha usando `for`.

```python
# Cria uma lista de quadrados
quadrados = [x**2 for x in range(10)]
print(quadrados)

# Cria uma lista com apenas números pares
pares = [x for x in range(10) if x % 2 == 0]
print(pares)
```

#### 11. `for` com Else
O `for` pode ter uma cláusula `else` que é executada quando o laço termina normalmente, ou seja, sem interrupção por `break`.

```python
for i in range(5):
    print(i)
else:
    print("Laço concluído sem interrupções.")

# Com break
for i in range(5):
    if i == 3:
        break
    print(i)
else:
    print("Este não será impresso porque o laço foi interrompido.")
```

#### 12. Iterando sobre um Arquivo
Você pode iterar sobre as linhas de um arquivo usando `for`.

```python
with open('arquivo.txt', 'r') as arquivo:
    for linha in arquivo:
        print(linha.strip())  # .strip() remove espaços em branco ao redor
```

Essas são as principais formas de utilizar o `for` em Python. Dependendo da situação, você pode escolher a abordagem que melhor se adapta ao problema que está resolvendo.

## 3. Simulação de `do while`
Como Python não tem um laço do while, você pode simular um usando um while loop com uma condição inicial verdadeira e uma verificação de condição no final.

```python
# Exemplo: Simulação de do while
contador = 5
while True:
    print(contador)
    contador -= 1
    if contador <= 0:
        break
```
